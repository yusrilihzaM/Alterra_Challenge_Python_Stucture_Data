```
def pengelompokanAngka(A):
  final=[]
  genap=[]
  ganjil=[]
  kelipatan=[]
  for i in A:
    if(i%3==0):
      kelipatan.append(i)
    elif(i%2==0):
      genap.append(i)
    else:
      ganjil.append(i)
  final.append(genap)
  final.append(ganjil)
  final.append(kelipatan)
  return final
print(pengelompokanAngka([2,4,6]))
print(pengelompokanAngka([1,2,3,4,5,6,7,8,9]))
print(pengelompokanAngka([100,151,122,99,111]))
```

```
def groupAnimals(Animal):
  hasil=[]
  while Animal:
    data_awal=Animal[0]
    for i in Animal:
      if i < data_awal:
         data_awal = i
    hasil.append(data_awal)
    Animal.remove(data_awal)
  return hasil
print(groupAnimals(["Ayam","Kucing","Bebek","Zebra","Katak"]))
```

```
def totalDigitRekursif(Angka):
    pecahin=[]
    for i in range(len(str(Angka))):
      pecahin.append(int(str(Angka)[i]))
    return(sum(pecahin))
print("512->",totalDigitRekursif(512))
print("512->",totalDigitRekursif(1542))
print("5->",totalDigitRekursif(5))
print("21->",totalDigitRekursif(21))
print("1111->",totalDigitRekursif(11111))
```
